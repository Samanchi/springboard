```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)

```

```python
NanFree['Year']=NanFree['Month'].apply(lambda t:t.year)
NanFree.head()
```

```python
NanMean=NanFree.groupby(by=['London_Borough','Year']).mean()
```

```python
NanMean.sample(15)
```

```python
NanMean=NanMean.reset_index()
```

```python
NanMean.head()
```

```python
def price_ratio(d):
    y2000=float(d['Average_price'] [d['Year']==2000])
    
    y2020=float(d['Average_price'] [d['Year']==2020])
    ratio=[y2020/y2000]
    return ratio
```

```python
price_ratio(NanMean[NanMean['London_Borough']=='Barking & Dagenham'])
```

```python
final={}
```

```python
for ratio in NanMean['London_Borough'].unique():
    borough=NanMean[NanMean['London_Borough']== ratio]
    final[ratio]=price_ratio(borough)
print(final)
```

```python
df=pd.DataFrame(final)
```

```python
df_tanspose=df.T
```

```python
df_tanspose.shape
```

```python
df_tanspose=df_tanspose.rename(columns={0:'Mean_price'})
```

```python
df_tanspose=df_tanspose.sort_values(by='Mean_price', ascending=False)
```
